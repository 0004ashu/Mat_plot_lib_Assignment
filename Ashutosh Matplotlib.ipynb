# Line plot


```python
Y = data.iloc[2:,1].values
E = data.iloc[2:,2].values
H = data.iloc[2:,3].values
Q = data.iloc[2:,4].values
X = data.iloc[2:,0]
plt.plot(X, Y)
```

# labels and titles


```python
plt.plot(X, Y)
plt.xlabel('District')
plt.ylabel('Active cases')
plt.title('District vs Active cases')
plt.show()
```

```python
plt.plot(X, E)
plt.xlabel('District')
plt.ylabel('Confirmed')
plt.title('District vs Confirmed cases')
plt.show()
```

```python
plt.plot(X, H)
plt.xlabel('District')
plt.ylabel('Recovered')
plt.title('District vs Recovered cases')
plt.show()
```

```python
plt.plot(X, Q)
plt.xlabel('District')
plt.ylabel('Deceased')
plt.title('District vs Deceased patients')
plt.show()
```

```python
plt.plot(X,Y)
plt.plot(X,E)
plt.plot(X,H)
plt.plot(X,Q)


plt.xlabel('District')
plt.ylabel('Active, Confirmed, Recovered and Deceased')
plt.title('DISTRICT vs Active, Confirmed, Recovered and Deceased ')
plt.show()
```

# legends for the graphs

```python
plt.plot(X, Y, label="Active cases", color ="purple")
plt.plot(X, E, label="Confirmed", color = "yellow")
plt.plot(X, H, label="Recovered " , color ="magenta")
plt.plot(X, Q, label="Deceased", color ="pink")
plt.xlabel('District')
plt.ylabel('Active, Confirmed, Recovered and Deceased')
plt.title('DISTRICT vs Active, Confirmed, Recovered and Deceased')
plt.legend() 
plt.show()
```

# Customisation

```python
plt.figure(figsize=(50,40))
plt.plot(X, Y, label="Active cases", color ="Purple")
plt.plot(X, E, label="Confirmed", color = "Yellow")
plt.plot(X, H, label="Recovered " , color ="Magenta")
plt.plot(X, Q, label="Deceased", color ="Pink")
plt.xlabel('District')
plt.ylabel('Active, Confirmed, Recovered and Deceased')
plt.title('DISTRICT vs Active, Confirmed, Recovered and Deceased')
plt.legend() 
plt.show()
```

```python
plt.figure(figsize=(20,30))
plt.plot(X, Y, label="Active cases", color ="purple", linewidth = 4, marker ='.' , markersize='20')
plt.plot(X, E, label="Confirmed", color = "yellow",linewidth = 4, marker ='x' , markersize='20')
plt.plot(X, H, label="Recovered " , color ="magenta", linewidth = 4, marker ='*' , markersize='20')
plt.plot(X, Q, label="Deceased", color ="pink", linewidth = 4, marker ='+' , markersize='20')
plt.xlabel('District')
plt.ylabel('Active, Confirmed, Recovered and Deceased')
plt.title('DISTRICT vs Active, Confirmed, Recovered and Deceased')
plt.legend() 
plt.show()
```

# Bar graph

```python
df = pd.DataFrame({
    'District': ['Ahmadnagar', 'Yavatmal', 'Washim', 'Solapur', 'Sindhudurg'],
    'active':[17,69,1,93,1],
    'Confirmed': [42,79,2,99,2],
    'Recovered':[23,10,1,0,1],
    'Deceased':[2,0,0,6,0]
    
    
})

ax = df.plot(x="District",y="active",kind="bar")
df.plot(x="District", y="Recovered", kind="bar", ax=ax, color="red")
```
